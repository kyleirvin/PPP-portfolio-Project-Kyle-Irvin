## PPP Data Consolidation

```python
#convert PPP files downloaded from (https://data.sba.gov/dataset/ppp-foia) into 1 file.
#The dataset comes in 12 files, consolidate into 1 for future use
```

```python
#Import libraries, set max columns
import pandas as pd
import glob2 as glob

pd.set_option('display.max_columns', 500)
```

```python
#This code consolidates all 12 files into 1 file using wildcard search to find all csv files in the designated folder,
#and then using concat to combine into 1 
ppp_df = pd.concat(map(pd.read_csv, glob.glob('*.csv')))
```
